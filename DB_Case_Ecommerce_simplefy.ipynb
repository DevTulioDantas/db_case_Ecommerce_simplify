Estudo de caso: E-COMMERCE

Desafio: Análise de Vendas e Clientes
Você deve criar um banco de dados para uma loja online e responder a algumas perguntas de negócio usando SQL.

Para o desafio consideramos um banco de dados com a seguinte estrutura:

*customer*(id_customer PK, customer_name, email, cellphone, register_date)

*products*(id_products PK, product_name, category, price)

*orders* (id_orders pk, id_products fk, id_customer fk, quantity, order_date, total_price, price_at_order)

adotamos no nosso banco o preço do produto salvo na tabela products, na tabela orders temos novamente o preço mas salvo em uma nova coluna chamada *price_at_order* para quando tiver alterações no preço na tabela product não alterar o preço dos orders ja registrados

portanto o nosso banco ficou assim:


image.png

```python
-- Tabela customer
CREATE TABLE customer (
    id_customer INT PRIMARY KEY,
    customer_name VARCHAR(255) NOT NULL,
    email VARCHAR(255) NOT NULL,
    cellphone VARCHAR(15),
    register_date DATE NOT NULL
);

-- Tabela products
CREATE TABLE products (
    id_products INT PRIMARY KEY,
    product_name VARCHAR(255) NOT NULL,
    category VARCHAR(100),
    price DECIMAL(10,2) NOT NULL
);

-- Tabela orders
CREATE TABLE orders (
    id_orders INT PRIMARY KEY,
    id_products INT NOT NULL,
    id_customer INT NOT NULL,
    quantity INT NOT NULL CHECK (quantity > 0),
    order_date DATE NOT NULL,
    total_price DECIMAL(10,2) GENERATED ALWAYS AS (quantity * price_at_order) STORED,
    price_at_order DECIMAL(10,2) NOT NULL,
    FOREIGN KEY (id_products) REFERENCES products(id_products),
    FOREIGN KEY (id_customer) REFERENCES customer(id_customer)
);
```

Considerando o banco pronto foram feitos os seguintes desafios:

Desafio 1:

Crie uma consulta que traga todos os clientes e os pedidos realizados por eles, mostrando as colunas:

Nome do cliente

Data do pedido

Quantidade

Preço do produto na data do pedido


```python
select
	customer_name, order_date, quantity, price_at_order
from orders o
join customer c on o.id_customer = c.id_customer
order by customer_name, o.order_date
```

2. Total de pedidos por cliente:
Crie uma consulta que calcule o total de pedidos realizados por cada cliente, mostrando:

Nome do cliente

Quantidade total de pedidos

Valor total gasto (somando os total_price dos pedidos)

```python
select
	customer_name, count(id_orders), sum(total_price)
from orders o
join customer c on o.id_customer = c.id_customer
group by customer_name
```

3. Produtos mais vendidos:
Crie uma consulta que traga os 10 produtos mais vendidos em termos de quantidade, incluindo:

Nome do produto,
Quantidade total vendida

```python
select
	sum(quantity) as Quantidade_total, product_name, sum(total_price) as valor_total
from orders o
join products p on o.id_products = p.id_products
group by product_name
order by sum(quantity) desc LIMIT 10;
```

4. Clientes que gastaram mais de R 500,00

Crie uma consulta que liste os clientes que gastaram mais de R$ 500,00 em total_price nos seus pedidos, incluindo:

Nome do cliente

Total gasto

```python
select
	customer_name, sum(total_price)
	from customer c
join orders o on o.id_customer = c.id_customer
group by customer_name
having sum(o.total_price) > 500
```

5. Listar produtos com vendas abaixo de 10 unidades:
Crie uma consulta para listar os produtos que venderam menos de 10 unidades no total, incluindo:

Nome do produto

Quantidade total vendida


```python
select
	product_name, sum(quantity) as total_quantity
	from products p
join orders o on o.id_products = p.id_products
group by product_name
having sum(quantity) < 10
order by total_quantity ASC
```

Pedidos em uma data específica:
Crie uma consulta que mostre todos os pedidos feitos no dia 2025-03-15, incluindo:

Nome do cliente

Nome do produto

Quantidade

Preço na data do pedido

```python
select
	customer_name, product_name, quantity, total_price, order_date, price_at_order
	from orders o
join products p on o.id_products = p.id_products
join customer c on o.id_customer = c.id_customer
where order_date = '2025-03-15'
```

7. Clientes que compraram mais de um produto:

Crie uma consulta que liste os clientes que compraram mais de um produto, incluindo:

Nome do cliente

Quantidade de produtos comprados (em termos de pedidos diferentes)


```python
select
	customer_name, sum(quantity)
	from orders o
join customer c on o.id_customer = c.id_customer
group by customer_name
```

8. Produtos vendidos em março de 2025:
Crie uma consulta que mostre todos os produtos vendidos no mês de março de 2025, incluindo:

Nome do produto

Quantidade vendida

Total de vendas (quantidade * preço)

```python
select
	product_name, sum(quantity), sum(total_price)
	from orders o
join products p on o.id_products = p.id_products
join customer c on o.id_customer = c.id_customer
WHERE order_date between '2025-03-01' and '2025-03-31'
group by product_name
```

10. Valor total dos pedidos por mês:

Crie uma consulta que calcule o valor total de pedidos feitos em cada mês de 2025, agrupando por mês e mostrando:

Mês (ano-mês)

Valor total de pedidos

```python
SELECT EXTRACT(MONTH FROM order_date) AS mes, sum(total_price)
FROM orders
GROUP BY mes;
```
